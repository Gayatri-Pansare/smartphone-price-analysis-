# Data Assessing
## Data quality issues are :- `completeness`, `validity`,`Accuracy`,`Consistency`

### Quality Issues

1. **model** - some brands are written diiferently like OPPO and LAVA in model column `consistency`
2. **price** - has unneccesary '₹' and "," between numbers `validity`
3. **Specification score** - has 'Spec score' after numbers `validity`
4. **Product name** - row '674' MTRJT is feature phone
5. **RAM** - row '539' and '275' has only ROM RAM is not given `completeness`
6. **Battery** - row '328','219','170','159','144','126','102','94','95','68','49','43', has no charger info `completeness`
7. **camera** row - 184,189,287,424,430,433,447,473,499,534,555,589,591,599,602,665,666,705,707,709,743,764,765,766,769,796,875,876,886,899,929,934,950,978,979  `validity`
8. **display** - sometimes frequency is not available `completeness`
9. row '496','660' = battery column has display info and disply column has camera info and camera column has memory card info and memory card column has OS info
11. **memory card** and **OS** - has scattered info
12. datatype of price and Specification score] is incorrect `validity`
13. There are 5 duplicate rows are present

### Tidiness Issues

1. **sim** - can be split into 3 cols has_5g, has_NFC, has_IR_Blaster
2. **ram** - can be split into 2 cols RAM and ROM
3. **processor** - can be split into processor name, cores and cpu speed.
4. **battery** - can be split into battery capacity, fast_charging_available
5. **display** - can be split into size, resolution_width, resolution_height and frequency
6. **camera** - can be split into front and rear camera
7. **card** - can be split into supported, extended_upto

## Creating new columns from "Camera" column

```python
def camera_extractor(text):
    if "Quad" in text:
        return 4
    elif "Triple" in text:
        return 3
    elif "Dual" in text:
        return 2
    elif 'Missing' in text:
        return 'Missing'
    else:
        return '1'
```
