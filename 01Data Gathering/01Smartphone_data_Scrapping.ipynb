```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
with open('smartprix.html','r',encoding='utf-8') as f:
    html_code=f.read()
```

```python
html_code[:1000]
```

```python
soup = BeautifulSoup(html_code)
```

```python
soup.text[:1000]
```

```python
containers = soup.find_all('div',attrs={'class':'sm-product has-tag has-features has-actions'})
len(containers)
```

```python
name_list = []
price_list =[]
spec_score_list = []
sim_list = []
processor_list = []
RAM_storage_list = []
battery_charger_list = []
display_list = []
camera_list = []
memory_card_list = []
OS_list = []


for c in containers:
    # Product name
    name = c.find('a' , attrs={'class':'name clamp-2'})
    if name is None:
        name_list.append(np.nan)
    else:
        name_list.append(name.text)

    # Product price
    price = c.find('span' , attrs= {'class':'price'})
    if price is None:
       price_list.append(np.nan)
    else:
        price_list.append(price.text)

    # Product spec_score
    spec_score = c.find('div' , attrs= {'class':'tags'})
    if spec_score is None:
        spec_score_list.append(np.nan)
    else:
        spec_score_list.append(spec_score.text)

    # Product features
    x = c.find('ul' , attrs= {'class':'sm-feat specs'}).find_all('li')

    # sim
    if x[0] is None:
        sim_list.append(np.nan)
    else:
        sim_list.append(x[0].text)

    # Processor
    if x[1] is None:
        processor_list.append(np.nan)
    else:
        processor_list.append(x[1].text)

    # RAM
    if x[2] is None:
        RAM_storage_list.append(np.nan)
    else:
        RAM_storage_list.append(x[2].text)

    # Battery
    if x[3] is None:
        battery_charger_list.append(np.nan)
    else:
        battery_charger_list.append(x[3].text)

    # display
    if x[4] is None:
        display_list.append(np.nan)
    else:
        display_list.append(x[4].text)

    # camera
    if x[5] is None:
        camera_list.append(np.nan)
    else:
        camera_list.append(x[5].text)

    # memory card
    if x[6] is None:
        memory_card_list.append(np.nan)
    else:
        memory_card_list.append(x[6].text)

    # OS
    if x[7] is None:
        OS_list.append(np.nan)
    else:
        OS_list.append(x[7].text)
    
```

```python
print(len(name_list))
print(len(price_list))
print(len(spec_score_list))
print(len(sim_list))
print(len(processor_list))
print(len(RAM_storage_list))
print(len(battery_charger_list))
print(len(display_list))
print(len(camera_list))
print(len(memory_card_list))
print(len(OS_list))
```

```python
df = pd.DataFrame({
    'Product name': name_list,
    'Price': price_list,
    'Specification Score': spec_score_list,
    'Sim': sim_list,
    'Processor':processor_list,
    'RAM and Storage':RAM_storage_list,
    'Battery':battery_charger_list,
    'Display':display_list,
    'Camera':camera_list,
    'Memory Card':memory_card_list,
    'OS':OS_list   
})
```
